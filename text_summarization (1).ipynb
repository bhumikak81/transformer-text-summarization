```python
from transformers import pipeline
```

```python
summarizer = pipeline("summarization", model="facebook/bart-large-cnn")
```

```python
input_text = input("Enter a paragraph you'd like to summarize:\n\n")
```

```python
summary = summarizer(input_text, max_length=60, min_length=20, do_sample=False)
```

```python
print("\n Original Text:\n", input_text)
print("\n Summarized Text:\n", summary[0]['summary_text'])
```
